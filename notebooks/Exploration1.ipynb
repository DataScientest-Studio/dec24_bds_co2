```python
import pandas as pd
```

```python
import chardet

# Lire un échantillon du fichier
with open('mars2014.csv', 'rb') as f:
    result = chardet.detect(f.read())  # Analyse les 100 000 premiers octets
    encoding = result['encoding']
    print(f"L'encodage détecté est : {encoding}")

# Lire le fichier avec l'encodage détecté
df = pd.read_csv('mars2014.csv', encoding='ISO-8859-1', sep=';')
```

```python
print(df.shape)
```

```python
def presentation_donnees(data):

    missing_values = data.isna().sum().sum()
    print("nombre de valeurs manquantes NaN :", missing_values)
    print("rate of missing value :", missing_values/data.size *100, "%")
    print("Number of rows:", data.shape[0])
    print("number of columns:", data.shape[1]," colonnes : ", data.columns)
    
    display(data.head())
    display(data.info())
    
```

```python
presentation_donnees(df)
```

```python
df.isna().sum()
```

```python
df_nona = df.drop(['Unnamed: 26',
       'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29','date_maj','ptcl','hcnox','nox','nox','hc','co_typ_1'], axis=1)
df_nona = df_nona.dropna()
presentation_donnees(df_nona)
```

```python
df_nona
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df_nona, x ='gamme' )
plt.xticks(rotation=45, fontsize=10, ha='right')  
plt.xlabel('Gamme', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Répartition des gammes', fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
sns.catplot(data=df_nona, x='gamme', y='co2', kind='box')
plt.xticks(rotation=45, fontsize=10, ha='right')  
plt.xlabel('Gamme', fontsize=12)
plt.ylabel('co2', fontsize=12)
plt.title('co2 par gammes', fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df_nona, x='gamme', y='co2')
plt.xticks(rotation=45, fontsize=10, ha='right')  
plt.xlabel('Gamme', fontsize=12)
plt.ylabel('co2', fontsize=12)
plt.title('co2 par gammes', fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
object_columns = df_nona.select_dtypes(include=['object']).columns
for col in object_columns:
    modified_col = df_nona[col].str.replace(',', '.', regex=True)
    if not modified_col.equals(df_nona[col]):
        df_nona[col] = pd.to_numeric(modified_col, errors='coerce')
        
df_nona[['masse_ordma_min', 'masse_ordma_max']] = df_nona[['masse_ordma_min', 'masse_ordma_max']].apply(
    pd.to_numeric, errors='coerce')
df_nona.info()
```

```python
numeric_df_nona = df_nona.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df_nona.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap des corrélations")
plt.show()
```
